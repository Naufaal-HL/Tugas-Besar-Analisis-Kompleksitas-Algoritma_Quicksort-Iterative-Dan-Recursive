### **Best Case Scenario Iterative Dan Recursive**

Dalam kasus yang paling seimbang, setiap kali kami melakukan partisi, kami membagi daftar menjadi dua bagian yang hampir sama. Ini berarti setiap panggilan rekursif memproses daftar setengah ukuran. Akibatnya, kita hanya dapat membuat log2n panggilan bersarang sebelum kita mencapai daftar ukuran 1. Ini berarti kedalaman pohon panggilan adalah log2n . Tetapi tidak ada dua panggilan pada tingkat yang sama dari pohon panggilan yang memproses bagian yang sama dari daftar asli; dengan demikian, setiap level panggilan hanya membutuhkan O(n) waktu bersama-sama (setiap panggilan memiliki beberapa overhead konstan, tetapi karena hanya ada O(n) panggilan di setiap level, ini dimasukkan dalam O(n) faktor). Hasilnya adalah algoritma hanya menggunakan waktu O(nlogn) .

## **Modelling Analisis Kompleksitas Waktu Algoritma**

```
# Mengimport Library Yang Dibutuhkan
import matplotlib.pyplot as plt
import seaborn as sns 
import pandas as pd
import numpy as np
import timeit
import sys
import os

# Set Jumlah Limit Maksimal Proses Rekursif
sys.setrecursionlimit(4000)
os.makedirs('data', exist_ok=True)

# Mendefinisikan Function Calculate Untuk Perhitungan Waktu Algoritma
def calculate(fname, start):
    stop = timeit.default_timer()
    execution_time = stop - start
    return execution_time
```

```
# Algoritma Quick Sort Rekursif
def quick_sort_recursion(S):
  def partition(S, low, high):
      i = (low - 1)       
      pivot = S[high]    

      for j in range(low, high):

          if S[j] <= pivot:

              i += 1
              S[i], S[j] = S[j], S[i]

      S[i + 1], S[high] = S[high], S[i + 1]
      return (i + 1)

  def quickSort(S, low, high):
      if low < high:

          pi = partition(S, low, high)

          quickSort(S, low, pi-1)
          quickSort(S, pi + 1, high)
  
  n = len(S)
  quickSort(S, 0, n - 1)
```

```
# Algoritma Quick Sort Iteratif
def quick_sort_iterative(S):
    def partition(S, l, h):
        i = (l - 1)
        x = S[h]

        for j in range(l, h):
            if S[j] <= x:

                i = i+1
                S[i], S[j] = S[j], S[i]

        S[i+1], S[h] = S[h], S[i+1]
        return (i+1)

    def quickSortIterative(S, l, h):
        size = h - l + 1
        stack = [0] * (size)

        top = -1
        top = top + 1
        stack[top] = l
        top = top + 1
        stack[top] = h

        while top >= 0:
            h = stack[top]
            top = top - 1
            l = stack[top]
            top = top - 1

            p = partition(S, l, h)

            if p-1 > l:
                top = top + 1
                stack[top] = l
                top = top + 1
                stack[top] = p - 1

            if p+1 < h:
                top = top + 1
                stack[top] = p + 1
                top = top + 1
                stack[top] = h
  
    n = len(S)
    quickSortIterative(S, 0, n-1)
```

```
# Menggenerate Angka Untuk Test Kompleksitas 
# Terdiri Atas Data Random, Urut Ke Terkecil, Dan Nilai Yang Sama
size = [1000, 1500, 2000, 2500, 3000, 3500]

for n in size:
    array = list(np.random.randint(1, 5000,  n))
    name = 'rand' + str(n) + '.dat'
    np.savetxt('data/'+name, array, fmt=['%i'])

for x in size:
    number = []

    for n in range(x, 0, -1):
        number.append(n)

    name = 'reversed' + str(x) + '.dat'
    np.savetxt('data/'+name, number, fmt=['%i'])

for x in size:
    number = []
    for _ in range(x):
        number.append(5)

    name = 'equal' + str(x) + '.dat'
    np.savetxt('data/' + name, number, fmt=['%i'])
```

```
# Melakukan Proses Perhitungan Waktu Berdasarkan Masing-Masing Test Case
qs = []

# Perulangan Untuk Quicksort Recursive Dengan Data Random
for n in size:
    name = 'rand' + str(n) + '.dat'

    for _ in range(3):
      data = np.loadtxt('data/' + name)
      start = timeit.default_timer()
      quick_sort_recursion(data)  
      qs.append([n, calculate('Quicksort Recursive', start), 'Quicksort Recursive Random'])

# Perulangan Untuk Quicksort Iterative Data Random
for n in size:
    name = 'rand' + str(n) + '.dat'

    for _ in range(3):
      data = np.loadtxt('data/' + name)
      start = timeit.default_timer()
      quick_sort_iterative(data)  
      qs.append([n, calculate('Quicksort Iterative', start), 'Quicksort Iterative Random'])
```

```
# Perulangan Untuk Quicksort Recursive Data Sama
for n in size:
    name = 'equal' + str(n) + '.dat'

    for _ in range(3):
      data = np.loadtxt('data/' + name)
      start = timeit.default_timer()
      quick_sort_recursion(data)  
      qs.append([n, calculate('Quicksort Recursive', start), 'Quicksort Recursive Equal'])
      
# Perulangan Untuk Quicksort Iterative Data Sama
for n in size:
    name = 'equal' + str(n) + '.dat'
      
    for _ in range(3):
      data = np.loadtxt('data/' + name)
      start = timeit.default_timer()
      quick_sort_iterative(data)  
      qs.append([n, calculate('Quicksort Iterative', start), 'Quicksort Iterative Equal'])
```

```
# Perulangan Untuk Quicksort Recursive Data Terurut Mengecil
for n in size:
    name = 'reversed' + str(n) + '.dat'

    for _ in range(3):
      data = np.loadtxt('data/' + name)
      start = timeit.default_timer()
      quick_sort_recursion(data)  
      qs.append([n, calculate('Quicksort Recursive', start), 'Quicksort Recursive Reversed'])

# Perulangan Untuk Quicksort Iterative Data Terurut Mengecil
for n in size:
    name = 'reversed' + str(n) + '.dat'

    for _ in range(3):
      data = np.loadtxt('data/' + name)
      start = timeit.default_timer()
      quick_sort_iterative(data)  
      qs.append([n, calculate('Quicksort Iterative', start), 'Quicksort Iterative Reversed'])
```

```
# Membuat Data Array Menjadi Dataframe Untuk Dapat Diolah Lebih Lanjut
qs_np = np.array(qs)
data = pd.DataFrame({'Jumlah Data': qs_np[:, 0], 'Kompleksitas Waktu': qs_np[:, 1], 'Kategori': qs_np[:, 2]})
```

```
# Melakukan Konversi Dan Pembulatan Hasil Perhitungan Kompleksitas Waktu
def float_qs(x):
  return round(float(x), 3)

data['Kompleksitas Waktu'] = data['Kompleksitas Waktu'].apply(float_qs)
```

```
# Memisahkan Data Berdasarkan Jenis Data Test Untuk Ditampilkan Pada Tabel
qs_reversed = qs_reversed.sort_values(by=['Jumlah Data', 'Kategori'])
qs_random = qs_random.sort_values(by=['Jumlah Data', 'Kategori'])
qs_equal = qs_equal.sort_values(by=['Jumlah Data', 'Kategori'])

# Mengubah Jenis Perhitungan Menjadi Array Untuk Diolah Perhitungan Kompleksitasnya
data_qs_reversed = [list(x) for x in qs_reversed.to_records(index=False)]
data_qs_random = [list(x) for x in qs_random.to_records(index=False)]
data_qs_equal = [list(x) for x in qs_equal.to_records(index=False)]
```

```
# Melakukan Perhitungan Rata-Rata Kompleksitas Waktu Dari Setiap Jenis Perhitungan
def sum_data(data_qs):
    def float_qs(x):
      return round(float(x), 3)

    qs_sum = []
    sum = 0
    c_jumlah = 0

    for n in data_qs:
        sum += n[1]
        c_jumlah += 1

        if c_jumlah == 3:
          qs_sum.append([n[0], float_qs(sum / 3), n[2]])
          sum = 0
          c_jumlah = 0
    
    return qs_sum

# Mengubah Data Menjadi Dataframe Untuk Disusun Pada Tabel Sesuai Jenis Perhitungan
def table_qs_sum(qs_sum):
    qs_sum = np.array(sum_data(qs_sum))
    qs_sum = pd.DataFrame({'Jumlah Data': qs_sum[:, 0], 'Rata-Rata Kompleksitas Waktu (s)': qs_sum[:, 1], 'Kategori': qs_sum[:, 2]})
    return qs_sum
```
